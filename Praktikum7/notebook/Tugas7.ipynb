```python
import pandas as pd

import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score, mean_squared_error
```

```python
#ubah kolom Mg ke float
df['Mg'] = pd.to_numeric(df['Mg'], errors='coerce')

#hilangkan baris yang ada missing value
df = df.dropna()
```

```python
# x = df[['b2', 'b3', 'b2', 'b1', 'Sigma_VV', 'Sigma_VH']]

# y = df['N']
```

```python
x = df[['P', 'K', 'Ca', 'Mg', 'Fe', 'Mn', 'Cu', 'Zn', 'B', 'b12', 'b11', 'b9', 'b8a', 'b8', 'b7', 'b6', 'b5', 'b4', 'b3', 'b2', 'b1', 'Sigma_VV', 'Sigma_VH', 'plia', 'lia', 'iafe', 'gamma0_vv', 'gamma0_vh', 'beta0_vv', 'beta0_vh']]

y= df['N']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
model = LinearRegression()

model.fit(X_train, y_train)
```

1. Pembersihan Data: Baris pertama mengubah kolom 'Mg' menjadi tipe data numerik (pd.to_numeric), dengan opsi errors='coerce' yang akan mengubah nilai yang tidak valid (non-numerik) menjadi Missing Value (NaN). Setelah itu, baris df = df.dropna() menghapus semua baris dalam DataFrame (df) yang mengandung nilai Missing Value (NaN) di kolom mana pun, memastikan data bersih dari entri yang rusak atau hilang.

2. Pemilihan Fitur (X) dan Target (Y): Kode ini kemudian mendefinisikan variabel independen (X) yang akan digunakan untuk memprediksi, dan variabel dependen (y) yang merupakan target prediksi. Terdapat dua blok pemilihan X dan y, namun pada akhirnya blok kedua yang panjang mendefinisikan X sebagai kumpulan semua fitur (unsur hara, band citra, dan variabel radar) dan y sebagai kolom 'N' (kemungkinan unsur hara Nitrogen) yang akan diprediksi.

3. Pembagian Data: Fungsi train_test_split membagi pasangan data X dan y menjadi empat bagian: X_train dan y_train untuk melatih model, serta X_test dan y_test untuk menguji performa model, dengan test_size=0.1 berarti 10% data dialokasikan untuk pengujian.

4. elatihan Model: Akhirnya, sebuah objek LinearRegression dibuat dan disimpan dalam variabel model, lalu model tersebut dilatih (dipelajari) menggunakan data pelatihan (X_train dan y_train) melalui perintah model.fit().

```python
y_pred = model.predict(X_test)

#evaluasi model

r2 = r2_score(y_test, y_pred)

mse = mean_squared_error(y_test, y_pred)

rmse = np.sqrt(mse)

print("R² Score:", r2)

print("RMSE:", rmse)
```

kode ini menggunakan model regresi linier yang sudah dilatih di langkah sebelumnya untuk membuat prediksi, lalu menghitung seberapa baik performa model tersebut. Pertama, y_pred = model.predict(X_test) meminta model memprediksi nilai target (y) berdasarkan data pengujian yang belum pernah dilihat (X_test). Kemudian, bagian evaluasi menghitung dua metrik penting:r2 = r2_score(y_test, y_pred): Menghitung $R^2$ Score (Koefisien Determinasi), yang mengukur proporsi variabilitas dalam variabel dependen (target) yang dapat dijelaskan oleh model. Nilai $R^2$ yang dicetak sebesar 0.755 menunjukkan model dapat menjelaskan sekitar 75.5% dari variasi data.mse = mean_squared_error dan rmse = np.sqrt(mse): Menghitung $MSE$ (Mean Squared Error), yang mengukur rata-rata kuadrat kesalahan antara nilai sebenarnya (y_test) dan nilai prediksi (y_pred), lalu menghitung $RMSE$ (Root Mean Squared Error) dengan mengakarkannya. Nilai $RMSE$ sebesar 0.194 menunjukkan rata-rata besarnya kesalahan prediksi model dalam unit yang sama dengan variabel target ('N').

```python
coeff = pd.DataFrame({
    "Fitur": X_train.columns,
    "Koefisien": model.coef_
})

print(coeff)
```

membuat sebuah tabel baru (DataFrame) bernama coeff yang digunakan untuk memvisualisasikan Koefisien Regresi dari model yang sudah dilatih. Tabel ini memiliki dua kolom: kolom "Fitur" yang berisi semua nama variabel independen (diambil dari X_train.columns), dan kolom "Koefisien" yang berisi nilai koefisien regresi yang dihitung oleh model (model.coef_). Nilai koefisien ini menunjukkan seberapa besar perubahan pada variabel target ('N') untuk setiap kenaikan satu unit pada fitur tertentu, dengan asumsi fitur lain konstan. Dengan mencetak tabel coeff, kita bisa melihat fitur mana yang memiliki pengaruh positif terbesar, pengaruh negatif terbesar, atau pengaruh yang sangat kecil terhadap prediksi model.

```python
import statsmodels.api as sm

X_sm = sm.add_constant(x)

# tambahkan intercept

model_ols = sm.OLS(y, X_sm).fit()

print(model_ols.summary())
```

Mengimpor Alat Statistik: Baris import statsmodels.api as sm mengimpor pustaka statsmodels, yang merupakan pilihan populer di kalangan ahli statistik karena output yang sangat detail.Menambahkan Intercept (Konstanta): Baris X_sm = sm.add_constant(X) secara eksplisit menambahkan kolom konstanta (yang berisi angka 1) ke variabel independen X. Dalam analisis regresi, kolom ini sangat penting karena merepresentasikan nilai intercept  pada persamaan regresi (nilai prediksi Y ketika semua nilai X adalah nol).Melatih Model OLS: Baris model_ols = sm.OLS(y, X_sm).fit() membuat dan melatih model OLS (Ordinary Least Squares), yang merupakan metode untuk Regresi Linier, menggunakan data target (y) dan data fitur yang sudah ditambahkan konstanta (X_sm).Menampilkan Ringkasan Statistik: Terakhir, print(model_ols.summary()) mencetak ringkasan statistik yang komprehensif dari hasil regresi. Ringkasan ini biasanya mencakup koefisien,  Score, nilai p-value untuk menguji signifikansi setiap fitur, dan berbagai metrik diagnostik lainnya.
