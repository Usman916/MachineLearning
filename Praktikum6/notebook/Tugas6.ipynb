```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# 1. LOAD DAN PRE-PROCESSING DATA
# ==============================================================================
# Memuat dataset
df = pd.read_csv(path + "/kanker.csv")

# Menghapus kolom yang tidak relevan: 'id' dan 'Unnamed: 32' (kolom kosong)
df = df.drop(['id', 'Unnamed: 32'], axis=1)

# Memisahkan Fitur (X) dan Target (y)
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']

# Encode label target (diagnosis: B/M) menjadi nilai numerik (0/1)
# B (Benign/Jinak) -> 0, M (Malignant/Ganas) -> 1
le = LabelEncoder()
y_encoded = le.fit_transform(y)
class_names = le.classes_ # Untuk pelaporan: ['B', 'M']
```

```python
# 2. PEMBAGIAN DATA
# ==============================================================================
# Membagi data menjadi 80% untuk pelatihan (training) dan 20% untuk pengujian (testing)
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42
)
```

```python
# 3. PELATIHAN MODEL SVM
# ==============================================================================
# Membuat dan melatih model Support Vector Classifier (SVC) dengan kernel linear
model = SVC(kernel='linear', random_state=42)
model.fit(X_train, y_train)
```

```python
# 4. PREDIKSI DAN EVALUASI
# ==============================================================================
y_pred = model.predict(X_test)

# Akurasi
accuracy = accuracy_score(y_test, y_pred)
print(f"Akurasi Model SVM: {accuracy * 100:.2f}%")

# Laporan Klasifikasi
report = classification_report(y_test, y_pred, target_names=class_names)
print("\nLaporan Klasifikasi:\n", report)
```

```python
# 5. VISUALISASI CONFUSION MATRIX
# ==============================================================================
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 4))
sns.heatmap(
    cm,
    annot=True, # Tampilkan nilai
    fmt='d',    # Format sebagai integer
    cmap='Blues',
    xticklabels=class_names,
    yticklabels=class_names
)
plt.title("Confusion Matrix (Diagnosis Kanker Payudara)")
plt.xlabel("Label Prediksi")
plt.ylabel("Label Sebenarnya")
plt.show()
```

```python
# 6. VISUALISASI DATA 2D (Contoh: radius_mean vs texture_mean)
# ==============================================================================
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    df['radius_mean'],
    df['texture_mean'],
    c=y_encoded,
    cmap='coolwarm',
    alpha=0.7
)
plt.xlabel('Radius Mean')
plt.ylabel('Texture Mean')
plt.title('Visualisasi Fitur (Radius vs Texture) Berdasarkan Diagnosis')
# Menambahkan legenda
plt.legend(*scatter.legend_elements(), loc="lower right", title="Diagnosis", labels=class_names)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# --- Asumsi Variabel dan Data (untuk memastikan kode berjalan mandiri) ---
# df, y_encoded, dan class_names diinisialisasi ulang
df = pd.read_csv(path + "/kanker.csv")
df = df.drop(['id', 'Unnamed: 32'], axis=1)
y = df['diagnosis']
le = LabelEncoder()
y_encoded = le.fit_transform(y)
class_names = le.classes_ # ['B', 'M']
df['DiagnosisEncoded'] = y_encoded
# -------------------------------------------------------------------------

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Warna untuk tiap kelas: Hijau untuk Jinak (Benign), Merah untuk Ganas (Malignant)
colors = ['green', 'red']

# Fitur yang akan digunakan untuk plot 3D
feature_x = 'radius_mean'
feature_y = 'texture_mean'
feature_z = 'perimeter_mean'

# Plot tiap diagnosis dengan warna berbeda
for i, name in enumerate(class_names):
    subset = df[df['DiagnosisEncoded'] == i]
    ax.scatter(
        subset[feature_x],
        subset[feature_y],
        subset[feature_z],
        color=colors[i],
        label=name,
        s=50 # Ukuran titik
    )

# Mengatur Label Sumbu
ax.set_xlabel(feature_x.replace('_', ' ').title())
ax.set_ylabel(feature_y.replace('_', ' ').title())
ax.set_zlabel(feature_z.replace('_', ' ').title())
ax.set_title('3D Visualisasi Diagnosis Kanker Payudara')
ax.legend()

plt.show()

# Menghapus kolom sementara
df = df.drop('DiagnosisEncoded', axis=1)
```
