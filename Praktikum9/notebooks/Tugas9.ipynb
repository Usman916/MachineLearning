#1 Membuat dan Membersihkan Data

```python
df = pd.read_csv(path + "/data.csv")
df = df.drop(['id', 'Unnamed: 32'], axis=1)
```

#2 Encoding Target

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
```

#3 Definisikan Features (X) dan Target (Y)

```python
X = df.drop('diagnosis', axis=1)
Y = df['diagnosis']
```

#4. Pembagian Data (Train & Test Split)Data dibagi menjadi $80\%$ untuk training dan $20\%$ untuk testing.

```python
# Split data: 80% train, 20% test
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

print(f"Jumlah data training: {len(X_train)}")
print(f"Jumlah data testing: {len(X_test)}")
```

#5. Implementasi dan Training Model Gaussian Naive Bayes Karena fitur-fitur (seperti radius_mean, texture_mean, dll.) adalah data kontinu (tipe float), model yang paling sesuai adalah Gaussian Naïve Bayes.

```python
# Inisialisasi model
nb_model = GaussianNB()

# Training model
nb_model.fit(X_train, Y_train)

# Prediksi pada data test
Y_pred = nb_model.predict(X_test)
```

#6. Evaluasi Model (Accuracy, Report, Confusion Matrix)
Evaluasi dilakukan menggunakan metrik utama dalam klasifikasi: Accuracy Score, Classification Report (Precision, Recall, F1-Score), dan Confusion Matrix.

```python
# Accuracy Score
accuracy = accuracy_score(Y_test, Y_pred)
print(f"\nAccuracy Score: {accuracy:.4f}")

# Classification Report
print("\nClassification Report:")
print(classification_report(Y_test, Y_pred, target_names=['Benign (0)', 'Malignant (1)']))

# Confusion Matrix
cm = confusion_matrix(Y_test, Y_pred)
print("\nConfusion Matrix:")
print(cm)
```

#7. Cross Validation (K-Fold)Mengikuti panduan praktikum, kita akan menggunakan 5-Fold Cross-Validation untuk menguji konsistensi model pada 5 subset data yang berbeda.

```python
from sklearn.model_selection import cross_val_score

# K-Fold Cross-Validation (cv=5)
cv_nb = cross_val_score(nb_model, X, Y, cv=5, scoring='accuracy')

print("\nNaive Bayes Cross Validation Accuracy (5-Fold):")
print("Scores:", cv_nb)
print("Mean Accuracy:", cv_nb.mean())
print("Std Deviation:", cv_nb.std())
```
